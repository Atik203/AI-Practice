```python
'''
Problem 2
take an input n from keyboard
now
 if n is odd: derive the sum 1^2 + 3^2 + 5^2 + .... + n^2
 if n is even: derive the sum 1^2 + 3^2 + 5^2 + ..... + (n+1)^2
 if n is anything else: print "ERROR"
'''
def sum_of_squares(n):
    total = 0
    for i in range(1, n + 1, 2):
        total += i ** 2
    return total

n = int(input("Enter a number: "))
if n % 2 == 0:
    print(sum_of_squares(n+1))
elif n % 2 == 1:
    print(sum_of_squares(n))
else:
    print("ERROR") 
```

```python
'''
Problem 6
Write a function that takes a string as input and returns the number of vowels in the string.
'''

def count_vowels(text):
    vowels= "aeiouAEIOU"
    counter = 0
    for char in text:
        if char in vowels:
            counter += 1
    return counter

text=input("Enter a string: ")
print(f"No of vowels in the string: {count_vowels(text)}")

```

```python
'''
Problem 7
Write a function that takes a list of strings and returns a new list containing the lengths of each string.
'''

def str_lengths(strings):
    string_lengths = []

    '''
    for string in strings:
        string_lengths.append(len(string))  // short method
    '''
    #regular method
    for i in range(len(strings)):
        string_lengths.append(len(strings[i]))
    return string_lengths

strings = ["apple", "banana", "cherry"]
print(f"Lengths of strings: {str_lengths(strings)}")
```

```python
'''
Problem 8
Write a Python function to check if a given string is a palindrome.
'''

def is_palindrome(s):
    s = s.lower()  
    return s == s[::-1]  # Explanation: Check if string is equal to its reverse
s = input("Enter a string: ")
if is_palindrome(s):
    print(f'"{s}" is a palindrome.')
else:
    print(f'"{s}" is not a palindrome.')
```

```python
'''
Problem 9
Write a Python function that takes a list of numbers and returns the largest and smallest number in the list.
'''
def find_largest_smallest(numbers):
    if not numbers:
        return None, None
    largest = max(numbers)
    smallest = min(numbers)
    return largest, smallest
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]
print(f"Largest: {find_largest_smallest(numbers)[0]}, Smallest: {find_largest_smallest(numbers)[1]}")
```

```python
'''
Problem 10
Write a Python function to calculate the factorial of a non-negative integer.
'''

def fact(n):
    if(n<0):
        print("Invalid")
    else:
        if(n==0 or n==1):
            return 1
        else:
            return n*fact(n-1)

x=int(input("Enter a number:"))
print(f"Factorial of {x}={fact(x)}")
    
```

```python
'''

Problem 11
Write a Python function to check if a number is prime.

'''
def isPrime(n):
    if(n<0):
        print("Enter a non-negative Number")
    else:
        is_prime=True
        for i in range(2,n):
            if n%i==0:
                is_prime=False
    return is_prime
num=int(input("Enter a number:"))
print(isPrime(num))

```

```python
'''
Problem 12
Write a Python function to generate the Fibonacci sequence up to a given number of terms.
'''

def fibonacci(n):
    fib_sequence = []
    a, b = 0, 1
    for i in range(n):
        fib_sequence.append(a)
        temp=a
        a=b
        b=a+temp
    return fib_sequence
n = int(input("Enter the number of terms: "))
print(f"Fibonacci sequence up to {n} terms: {fibonacci(n)}")
```

```python
'''
Problem 13
Write a Python function to merge two dictionaries.
'''
def merge_dicts(dict1, dict2):
    if not isinstance(dict1, dict) or not isinstance(dict2, dict):
        print("ERROR")
        return
    merged_dict = dict1.copy()  
    merged_dict.update(dict2) #adding dict2 to dict1 
    return merged_dict

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}    
merged = merge_dicts(dict1, dict2)
print(f"Merged Dictionary: {merged}")

```

```python
'''
Problem 1
input 2 numbers, r (for number of rows) and c (for number of columns)
now for a 2D numpy array of shape (r, c), take the input for each element
now reshape the array to the shape (1, rc) and print the array

'''
import numpy as np
r=int(input("Enter number of rows: "))
c=int(input("Enter number of columns: "))
array=np.zeros((r, c), dtype=int)
for i in range(r):
    for j in range(c):
        array[i][j] = int(input(f"Enter value for element [{i}][{j}]: "))
print(array)
reshaped_array = array.reshape(1, r * c)
print(reshaped_array)
```

```python
'''

# Problem 1
# Plot a piecewise function  where y=x^2 for 0≤𝑥<4, y=16 for 4≤𝑥<8, and 𝑦=(12−𝑥)^2, for 8≤x≤12. Use Python's matplotlib library to create this plot."

'''
import matplotlib.pyplot as plt

x=[]
y=[]

for i in range(0,4):
    x.append(i)
    y.append(i**2)

for i in range(4,8):
    x.append(i)
    y.append(16)
for i in range(8,13):
    x.append(i)
    y.append((12-i)**2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph for function: y=x^2 for 0≤x<4, y=16 for 4≤x<8, and y=(12−x)^2, for 8≤x≤12")
plt.grid(True)
plt.show()
```

```python
'''
Problem 6
Create a subplot with two plots: a line plot of y = cos(x) and a scatter plot of random data.
'''
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y_cos = np.cos(x)
y_random = np.random.rand(100)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(x, y_cos)
plt.title('y = cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.scatter(x, y_random)
plt.title('Scatter Plot of Random Data')
plt.xlabel('x')
plt.ylabel('Random Values')
plt.show()
```
